```python
import pandas as pd
import altair as alt
```

```python
shots19 = pd.read_csv('shots-2019.csv')
shots19['distance'] = shots19['distance'].str.replace('ft', '').astype(int)
avg19 = shots19.groupby('team')['distance'].mean().reset_index()
avg19 = avg19.rename(columns={'distance':'2019'})
```

```python
total = pd.merge(avg19, avg17, on='team', how='outer'
            ).merge(avg15, on='team', how='outer'
                    ).merge(avg13, on='team', how='outer'
                            ).merge(avg11, on='team', how='outer'
                                   ).merge(avg09, on='team', how='outer'
                                          ).merge(avg07, on='team', how='outer'
                                                 ).merge(avg05, on='team', how='outer'
                                                        ).merge(avg03, on='team', how='outer'
                                                               ).merge(avg01, on='team', how='outer')
total.head()
```

```python
total.loc[2] = total.loc[2].combine_first(total.loc[30])
total = total.drop(total[total['team'] == 'New Jersey'].index, axis=0)
```

```python
total.loc[20] = total.loc[20].combine_first(total.loc[31])
total = total.drop(total[total['team'] == 'Seattle'].index, axis=0)
```

```python
total = total.drop(total.index[30])
```

```python
melted_teams = pd.melt(total, id_vars=['team'], var_name='Year', value_name='avg_distance')

melted_teams = melted_teams.sort_values(by=['team', 'Year']).reset_index(drop=True)

melted_teams = melted_teams.round(3)
```

```python
options = [None] + total['team'].unique().tolist()
labels = ['All'] + total['team'].unique().tolist()

input_dropdown = alt.binding_select(options = options,
                                    labels = labels,
                                    name = 'Team: ')

selection = alt.selection_point(fields = ['team'], bind = input_dropdown)

chart =    alt.Chart(melted_teams).mark_line().encode(
            x = 'Year:N',
            y = alt.Y('avg_distance:Q', title = 'Average Distance (ft)', scale=alt.Scale(domain=[9, 16])),
            color = alt.condition(selection, alt.Color('team:N'), 
                                  alt.value('white')),
            tooltip = alt.Tooltip(['team:N', 'Year:N', 'avg_distance:Q'])
        ).add_params(
            selection
        ).transform_filter(selection).properties(title = 'Average Shooting Distances (2001-2019)',
                    width = 500,
                    height = 300)

chart

#chart.save('shotcomparison.html')
```

```python
winning_team_df = shots19[shots19['team'] == shots19['winner']]
losing_team_df = shots19[shots19['team'] == shots19['loser']]

alt.data_transformers.disable_max_rows()

shots19["outcome"] = shots19["play"].apply(lambda x: str(x).split("<br>")[1].split()[-5]) 
shots19["attempt"] = shots19["play"].apply(lambda x: str(x).split("<br>")[1].split()[-4])

wins = alt.Chart(winning_team_df).mark_boxplot().encode(
    y = alt.Y('distance:Q', title='Shot Distance', scale=alt.Scale(domain=[0, 40])),
    x = alt.X('outcome:N', title = 'Shot Outcome')).properties(title='Shot Distance When Teams Win')

loss = alt.Chart(losing_team_df).mark_boxplot(color='red').encode(
    y = alt.Y('distance:Q', title='Shot Distance', scale=alt.Scale(domain=[0, 40])),
    x = alt.X('outcome:N', title = 'Shot Outcome')).properties(title='Shot Distance When Teams Lose')

boxplot = wins | loss
boxplot

#boxplot.save('boxplot.html')
```

```python
bucks = shots19[shots19['team'] == 'Milwaukee']
cavs = shots19[shots19['team'] == 'Cleveland']
bucks_shots = bucks.groupby(['attempt', 'outcome']).size().reset_index(name='count')
cavs_shots = cavs.groupby(['attempt', 'outcome']).size().reset_index(name='count')

bucks_chart = alt.Chart(bucks_shots).mark_bar().encode(
    x = alt.X('attempt:N', title='Shot Type'),
    y = alt.Y('count:Q', title='Number of Shots', scale=alt.Scale(domain=[0, 2250])),
    column = alt.Column('outcome:N', title = ''),
    tooltip=alt.Tooltip(['attempt:N', 'outcome:N', 'count:Q'])).properties(title='Milwaukee Bucks Shots',
                                                                          width = 100, height = 200)

cavs_chart = alt.Chart(cavs_shots).mark_bar().encode(
    x = alt.X('attempt:N', title='Shot Type'),
    y = alt.Y('count:Q', title='Number of Shots', scale=alt.Scale(domain=[0, 2250])),
    column = alt.Column('outcome:N', title = ''),
    tooltip=alt.Tooltip(['attempt:N', 'outcome:N', 'count:Q'])).properties(title='Cleveland Cavaliers Shots',
                                                            width = 100, height = 200)

c = bucks_chart | cavs_chart

c

#c.save('bucksvcavs.html')
```
